```python
import requests                     # For sending HTTP requests to fetch web page content
from bs4 import BeautifulSoup       # For parsing and extracting data from HTML/XML
import pandas as pd                 # For organizing and analyzing tabular data
from urllib.parse import urljoin    # For combining base and relative URLs into absolute URLs
```

```python
# URL of the page
link = 'http://www.hubertiming.com/results/2017GPTR10K'

# Step 1: Fetch and parse the webpage
response = requests.get(link)
response.raise_for_status()
soup = BeautifulSoup(response.text, 'html.parser')
tables = soup.find_all('table')
```

### Exercise 1: Get the title of the page

```python
# Function for Exercise 1: Get the title of the page
def get_page_title(soup):
    """
    Extracts and prints the title of the web page from the BeautifulSoup object.

    Returns:
        str: The title of the web page as a string.
    """
    title = soup.title.string
    print("Exercise 1: Page Title")
    return title

title = get_page_title(soup)
print(title)
```

### Exercise 2: Print the page as text

```python
# Function for Exercise 2: Print the page as text
def print_page_text(soup):
    """
    Prints the complete text content of a web page.

    Returns:
        None
    """
    print("\nExercise 2: Page Text")
    text = soup.get_text()
    print(text)

print_page_text(soup)   
```

### Exercise 3: Print out the links present on the page

```python
# Function for Exercise 3: Print out the links present on the page
def extract_links(soup, base_url):
    """
    Extracts and prints all the links from a web page, resolving relative URLs to absolute URLs.

    Returns:
        list: A list of absolute URLs present on the web page.
    """
    print("\nExercise 3: Links on the Page")
    links = [urljoin(base_url, a['href']) for a in soup.find_all('a', href=True)]
    
    print("There are %d links" % len(links))
    for link in links:
        print(" - %s" % link)
    
    return links

def filter_mailto_links(links):
    """
    Filters out links that start with 'mailto:' from a list of links.

    Returns:
        list: A list of filtered URLs excluding 'mailto:' links.
    """
    print("\nRemoving mailto links...")
    filtered_links = [link for link in links if not link.startswith('mailto:')]
    
    print("There are %d non-mailto links" % len(filtered_links))
    for link in filtered_links:
        print(" - %s" % link)
    
    return filtered_links

links = extract_links(soup, link) 
filtered_links = filter_mailto_links(links) 
```

### Exercise 4: Print the first 10 rows of each table for sanity check

```python
# Function for Exercise 4: Print the first 10 rows for sanity check
def print_first_10_rows(tables):
    """
    Prints the first 10 rows of each table on the web page

    Returns:
        None
    """
    print("\nExercise 4: First 10 Rows of Each Table (Excluding Headers)")
    
    for table_index, table in enumerate(tables):
        print(f"\nTable {table_index + 1}:")
        
        # Extract all rows
        rows = table.find_all('tr')
        if not rows:
            print("  No rows found in this table.")
            continue

        valid_rows = []
        for row in rows:
            cells = row.find_all('td')  # Exclude rows with <th> elements (headers)
            if cells:  # Only include rows with <td> elements
                valid_rows.append([cell.get_text(strip=True) for cell in cells])

        # Check if table has valid rows
        if not valid_rows:
            print("  No valid data rows found in this table.")
            continue

        # Print the first 10 rows or fewer if table has less than 10
        for i, row in enumerate(valid_rows[:10]):
            print(f"  Row {i + 1}: {row}")
            
print_first_10_rows(tables) 
```

### Exercise 5: Convert data to pd.DataFrame() format

```python
# Function for Exercise 5: Convert data to pd.DataFrame() format
def convert_to_dataframe(tables):
    """
    Converts HTML tables into pandas DataFrames, excluding empty rows.

    Returns:
        list: A list of pandas DataFrames, one for each table.
    """
    print("\nExercise 5: Convert Tables to DataFrames (Without Headers)")
    dataframes = []

    for table_index, table in enumerate(tables):
        # Extract data rows
        data = []
        for row in table.find_all('tr'):
            cells = row.find_all('td')
            if cells:  # Only include rows with <td> elements
                row_data = [cell.get_text(strip=True) for cell in cells]
                if any(row_data):  # Ensure the row is not completely empty
                    data.append(row_data)

        # Create DataFrame
        df = pd.DataFrame(data)  # Create a DataFrame without column names
        dataframes.append(df)

        # Debug Output
        print(f"\nTable {table_index + 1} DataFrame:")
        print(df.head())

    return dataframes

dataframes = convert_to_dataframe(tables)
```

### Exercise 6: Extract text between HTML tags for the table header and assign headers to the table.

```python
# Function for Exercise 6: Extract and assign headers
def extract_and_assign_headers(tables, dataframes):
    """
    Extracts headers from each table on the web page and assigns them to corresponding DataFrames.

    Returns:
        list: A list of pandas DataFrames with headers assigned.
    """
    print("\nExercise 6: Extract Table Headers and Assign to DataFrames")
    updated_dataframes = []

    for table_index, (table, df) in enumerate(zip(tables, dataframes)):
        # Extract headers
        headers = [header.get_text(strip=True) for header in table.find_all('th')]
        print(f"\nTable {table_index + 1} Headers: {headers}")

        # Assign headers to the corresponding DataFrame if headers are available
        if headers:
            df.columns = headers
        updated_dataframes.append(df)

        print(f"Updated DataFrame for Table {table_index + 1}:")
        print(df.head())

    return updated_dataframes

dataframes = extract_and_assign_headers(tables, dataframes)  # Assign headers
```

### Exercise 7: Handle Missing Values

```python
# Function for Exercise 7: Handle missing values
def handle_missing_values(dataframes):
    """
    Identifies and handles missing values in each DataFrame by replacing them with 'N/A'.

    Returns:
        None
    """
    print("\nExercise 7: Handle Missing Values")
    for table_index, df in enumerate(dataframes):
        print(f"\nProcessing Table {table_index + 1}:")
        num_null_values = df.isnull().sum().sum()  # Total count of NaN values
        num_empty_strings = (df == '').sum().sum()  # Total count of empty strings

        print(f"Total Null Values (NaN): {num_null_values}")
        print(f"Total Empty Strings: {num_empty_strings}")

        if num_null_values > 0 or num_empty_strings > 0:
            print("Missing values detected. Replacing with 'N/A'.")
            df.replace('', 'N/A', inplace=True)
            df.fillna('N/A', inplace=True)
        else:
            print("No missing values detected.")
        print("DataFrame Preview After Handling Missing Values:")
        print(df.head())

handle_missing_values(dataframes) 
```
